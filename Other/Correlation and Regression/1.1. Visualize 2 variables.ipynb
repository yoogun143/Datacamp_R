```r
library(openintro)
library(ggplot2)
library(dplyr)
```

## SCATTER PLOT

```r
# Mammals scatterplot
ggplot(mammals, aes(BodyWt, BrainWt)) +
  geom_point()
```

```r
# Scatterplot with coord_trans()
ggplot(data = mammals, aes(x = BodyWt, y = BrainWt)) +
  geom_point() + 
  coord_trans(x = "log10", y = "log10")
```

```r
# Scatterplot with scale_x_log10() and scale_y_log10()
ggplot(data = mammals, aes(x = BodyWt, y = BrainWt)) +
  geom_point() +
  scale_x_log10() + scale_y_log10()
```

## BOXPLOT AS DISCRETIZED SCATTERPLOT

```r
# Boxplot of weight vs. weeks
ggplot(data = ncbirths, 
       aes(x = cut(weeks, breaks = 5), y = weight)) +
  geom_boxplot()
```

## OUTLIER

```r
# Original scatter plot
ggplot(mlbBat10, aes(x = OBP, y = SLG)) +
  geom_point()
```

```r
# Scatterplot of SLG vs. OBP
mlbBat10 %>%
  filter(AB >= 200) %>%
  ggplot(aes(x = OBP, y = SLG)) +
  geom_point()
```

```r
# Identify the outlying player
mlbBat10 %>%
  filter(AB >= 200, OBP < 0.2)
```

## CORRELATION

```r
# Compute correlation
ncbirths %>%
  summarize(N = n(), r = cor(mage, weight))
```

```r
# Compute correlation for all non-missing pairs
ncbirths %>%
  summarize(N = n(), r = cor(weeks, weight, use = "pairwise.complete.obs"))
```
